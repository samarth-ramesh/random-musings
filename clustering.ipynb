```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
with open("./data.txt", "r") as fp:
    data = fp.readlines()
    data = tuple(map(lambda d : d.split(), data))
    data = tuple(map(lambda d: tuple(map(lambda x: int(x), d)), data))
arr = np.array(data)
# arr = np.transpose(data)
```

```python
transformed = arr.transpose()
plt.scatter(transformed[0], transformed[1])
```

```python
km = KMeans(n_clusters=50, init='random')
arr2 = km.fit_predict(arr)
```

```python
clusters = {}

for i in range(len(arr)):
    if not (arr2[i] in clusters):
        clusters[arr2[i]] = ([], [])
    clusters[arr2[i]][0].append(arr[i][0])
    clusters[arr2[i]][1].append(arr[i][1])
```

```python
for v in clusters.values():
    plt.scatter(v[0], v[1])    
```
